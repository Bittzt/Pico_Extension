```python
Read the extension's snippet JSON files and load the snippet entries for analysis.</VSCode.Cell>
<VSCode.Cell language="python">import json
from pathlib import Path

root = Path(r'c:\Users\kriti\OneDrive\Desktop\Project real\pico-helper')
snippet_files = list(root.glob('snippets/*.json'))
print('Found snippet files:', [p.name for p in snippet_files])

snippets = {}
for path in snippet_files:
    snippets[path.stem] = json.loads(path.read_text(encoding='utf-8'))
    print(path.name, 'entries:', len(snippets[path.stem]))

list(snippets.keys())
```
